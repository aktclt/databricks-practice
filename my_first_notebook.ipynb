# Serverless Databricks Connect

This notebook uses **only serverless** compute (`DatabricksSession.builder.serverless(True)`). No cluster ID.

**Kernel:** Python **3.11** — select `…/databricks-practice/.venv311/bin/python`. Serverless Connect in this setup expects that interpreter and a recent `databricks-connect`.

```python
import sys

if sys.version_info < (3, 11):
    raise RuntimeError(
        "Select kernel: .venv311 (Python 3.11). Current: " + sys.version.split()[0]
    )

from databricks.connect import DatabricksSession

spark = DatabricksSession.builder.serverless(True).getOrCreate()
spark
```

```python
spark.sql("SELECT current_user() AS user, current_timestamp() AS ts").show(truncate=False)
```

```python
df = spark.read.table("samples.nyctaxi.trips")
df.filter(df.trip_distance > 10).limit(5).show(5)
```
